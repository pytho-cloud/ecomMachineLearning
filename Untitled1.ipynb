```python
import pandas as pd 
import numpy as np

books = pd.read_csv("Books.csv")
users = pd.read_csv("Users.csv")
ratings = pd.read_csv("Ratings.csv")
ratings
```

```python
ratings_with_name = ratings.merge(books,on = "ISBN")
ratings_with_name =ratings_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()

nums = ratings_with_name.rename(columns = {"Book-Rating" : "Num-Rating"})
nums
```

```python
avg_sums = ratings.merge(books,on = "ISBN")
avg_sums = avg_sums.groupby("Book-Title")["Book-Rating"].mean().reset_index()
avg_sums = avg_sums.rename(columns = {"Book-Rating" : "Avg-Rating"})
avg_sums
```

```python
# Here I merge sums and avg of ratings and take that values whose nums of rating greater than 250
popular_df = nums.merge(avg_sums , on = "Book-Title")
popular_df = popular_df[popular_df["Num-Rating"] >= 250].sort_values("Avg-Rating" ,ascending = False)
popular_df = popular_df.merge(books , on ="Book-Title").drop_duplicates(["Book-Title"])[["Book-Title","Book-Author","Num-Rating","Avg-Rating",]]
popular_df.head(50)
```
